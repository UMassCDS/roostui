```python
# merge files for the same station-year
import os

BATCHES = ["texas_bats_v3"]
STATIONS = ["KSJT", "KGRK", "KDFX", "KEWX"]
YEARS = [str(i) for i in range(2000, 2021)]

for batch in BATCHES:
    scans = {(station, year): [] for station in STATIONS for year in YEARS}
    tracks = {(station, year): [] for station in STATIONS for year in YEARS}
    scans_first_line = None
    tracks_first_line = None

    for file in os.listdir(f"../data/{batch}_"):
        if file.startswith("scans"):
            if (file[6:10], file[11:15]) in scans:
                scans[(file[6:10], file[11:15])].append(file)
                if scans_first_line is None:
                    scans_first_line = open(
                        f"../data/{batch}_/{file}", "r"
                    ).readlines()[0]
                else:
                    assert (
                        scans_first_line
                        == open(f"../data/{batch}_/{file}", "r").readlines()[0]
                    )
        elif file.startswith("tracks"):
            if (file[7:11], file[12:16]) in tracks:
                tracks[(file[7:11], file[12:16])].append(file)
                if tracks_first_line is None:
                    tracks_first_line = open(
                        f"../data/{batch}_/{file}", "r"
                    ).readlines()[0]
                else:
                    assert (
                        tracks_first_line
                        == open(f"../data/{batch}_/{file}", "r").readlines()[0]
                    )

    assert scans_first_line is not None and tracks_first_line is not None
    os.makedirs(f"../data/{batch}", exist_ok=True)

    for (station, year), files in scans.items():
        lines = [scans_first_line]
        for file in sorted(files):
            lines.extend(open(f"../data/{batch}_/{file}", "r").readlines()[1:])
        with open(
            f"../data/{batch}/scans_{station}_{year}0101_{year}1231.txt", "w"
        ) as f:
            f.writelines(lines)

    for (station, year), files in tracks.items():
        lines = [tracks_first_line]
        for file in sorted(files):
            lines.extend(open(f"../data/{batch}_/{file}", "r").readlines()[1:])
        with open(
            f"../data/{batch}/tracks_{station}_{year}0101_{year}1231.txt", "w"
        ) as f:
            f.writelines(lines)
```

```python
# merge files for the same station-year
import os

batch = "birdwatcher_202210_v2"

scans = {}
tracks = {}
for file in os.listdir(f"../data/{batch}_"):
    if file.startswith("scans"):
        if (file[6:10], file[11:15]) in scans:
            scans[(file[6:10], file[11:15])].append(file)
        else:
            scans[(file[6:10], file[11:15])] = [file]
    elif file.startswith("tracks"):
        if (file[7:11], file[12:16]) in tracks:
            tracks[(file[7:11], file[12:16])].append(file)
        else:
            tracks[(file[7:11], file[12:16])] = [file]

os.makedirs(f"../data/{batch}", exist_ok=True)

for (station, year), files in scans.items():
    assert len(files) > 0
    for i, file in enumerate(sorted(files)):
        if i == 0:
            lines = [line for line in open(f"../data/{batch}_/{file}", "r").readlines()]
        else:
            lines.extend(
                [line for line in open(f"../data/{batch}_/{file}", "r").readlines()[1:]]
            )
    with open(f"../data/{batch}/scans_{station}_{year}0101_{year}1231.txt", "w") as f:
        f.writelines(lines)

for (station, year), files in tracks.items():
    assert len(files) > 0
    for i, file in enumerate(sorted(files)):
        if i == 0:
            lines = [line for line in open(f"../data/{batch}_/{file}", "r").readlines()]
        else:
            lines.extend(
                [line for line in open(f"../data/{batch}_/{file}", "r").readlines()[1:]]
            )
    with open(f"../data/{batch}/tracks_{station}_{year}0101_{year}1231.txt", "w") as f:
        f.writelines(lines)
```
