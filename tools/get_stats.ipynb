```python
# Number of scans
import os
from collections import defaultdict

DIR = "/scratch2/wenlongzhao/roostui/data/all_stations_v2_screened"

valid_station_year = set()
for file in sorted(os.listdir(DIR)):
    if file.startswith("roost_labels"):
        valid_station_year.add((file.split("_")[2], file.split("_")[3][:4]))
assert len(valid_station_year) == 246

n_scans = defaultdict(int)
all_n_scans = 0
for file in sorted(os.listdir(DIR)):
    if not file.startswith("scans"):
        continue
    if (file.split("_")[1], file.split("_")[2][:4]) not in valid_station_year:
        continue
    
    lines = open(os.path.join(DIR, file), "r").readlines()
    
    n = 0
    for line in lines[1:]:
        if int(line.split(",")[1][4:6]) < 11:
            n += 1
    n_scans[file.split("_")[1]] = n_scans[file.split("_")[1]] + n
    all_n_scans += n

for station in n_scans:
    print(station, n_scans[station])
print("Total", all_n_scans)
```

```python
# fields are:
#       0 track_id, 1 filename, 2 from_sunrise, 3 det_score, 4 x, 5 y, 6 r, 7 lon, 8 lat, 9 radius,
#       10 local_time, 11 station, 12 date, 13 time, 14 local_date, 15 length,
#       16 tot_score, 17 avg_score, 18 viewed, 19 user_labeled, 20 label, 21 original_label,
#       22 notes: 'LARGE', 'nr', 'long', 'large', 'rn', 'shrinks', 'shrink',
#       23 day_notes: 'pap', 'psp', 'weather', '2', 'ap', 'AP', 'miss', 'cluster', 'clusters',
import copy

YEARS = range(2000, 2021)
INIT = {
    "high_conf_dets": 0,
    "other_dets": 0,
    "other_dets_screened_as_roosts": 0,
    "high_conf_tracks": 0,
    "other_tracks": 0,
    "other_tracks_screened_as_roosts": 0,

    "swallow-roost": [set(), 0], # set for track_id, number of detections in these tracks
    "weather-roost": [set(), 0],
    "AP-roost": [set(), 0],
    "unknown-noise-roost": [set(), 0],
    "duplicate": [set(), 0],
    "bad-track": [set(), 0],
    "non-roost": [set(), 0],
}

stations = {"ALL": copy.deepcopy(INIT)}
for file in sorted(os.listdir(DIR)):
    if not file.startswith("roost_labels"):
        continue
        
    station = file.split("_")[2]
    year = file.split("_")[3][:4]
    assert int(year) in YEARS
    if station not in stations:
        stations[station] = copy.deepcopy(INIT)
    
    annotations = [annotation.strip().split(",") for annotation in open(os.path.join(DIR, file), "r").readlines()[1:]]
    tracks = {}
    for annotation in annotations:
        if int(annotation[10][4:6]) < 6 or int(annotation[10][4:6]) > 10: # check local month
            continue
        
        if annotation[0] not in tracks:
            tracks[annotation[0]] = [0, 0, 0, False, False]
                # sum score, num dets have score, num dets, has det w/ 0.5 score, screened as roosts
        
        # system predicted
        annotation[3] = float(annotation[3]) # det_score
        if annotation[3] > 0 and annotation[3] < 1:
            tracks[annotation[0]][0] = tracks[annotation[0]][0] + annotation[3] # sum score
            tracks[annotation[0]][1] = tracks[annotation[0]][1] + 1 # num dets have score
        tracks[annotation[0]][2] = tracks[annotation[0]][2] + 1 # num dets
        if annotation[3] >= 0.5: # has det w/ 0.5 score
            tracks[annotation[0]][3] = True
        
        # human screened
        if annotation[23].lower() in ["weather", "ap"]:
            stations[station]["non-roost"][0].add(annotation[0])
            stations[station]["non-roost"][1] += 1
            stations["ALL"]["non-roost"][0].add(annotation[0])
            stations["ALL"]["non-roost"][1] += 1
        else:
            stations[station][annotation[20]][0].add(annotation[0])
            stations[station][annotation[20]][1] += 1
            stations["ALL"][annotation[20]][0].add(annotation[0])
            stations["ALL"][annotation[20]][1] += 1
            if annotation[20] in ["swallow-roost", "weather-roost", "AP-roost", "unknown-noise-roost"]:
                tracks[annotation[0]][4] = True
    
    for track_id, track in tracks.items():
        if track[0] / track[1] >= 0.15 and track[2] >= 2 and track[3]:
            stations[station]["high_conf_dets"] += track[2]
            stations["ALL"]["high_conf_dets"] += track[2]
            stations[station]["high_conf_tracks"] += 1
            stations["ALL"]["high_conf_tracks"] += 1
        else:
            stations[station]["other_dets"] += track[2]
            stations["ALL"]["other_dets"] += track[2]
            stations[station]["other_tracks"] += 1
            stations["ALL"]["other_tracks"] += 1
            if track[4]:
                stations[station]["other_dets_screened_as_roosts"] += track[2]
                stations["ALL"]["other_dets_screened_as_roosts"] += track[2]
                stations[station]["other_tracks_screened_as_roosts"] += 1
                stations["ALL"]["other_tracks_screened_as_roosts"] += 1

for station in stations:
    print(
        station,
        f"{stations[station]['high_conf_dets']} & {stations[station]['other_dets']} & "\
        f"{stations[station]['other_dets_screened_as_roosts']} & "\
        f"{stations[station]['high_conf_tracks']} & {stations[station]['other_tracks']} & "\
        f"{stations[station]['other_tracks_screened_as_roosts']}"\
    )
print()
for station in stations:
    print(
        station,
        f"{stations[station]['swallow-roost'][1]}/{len(stations[station]['swallow-roost'][0])} & "\
        f"{stations[station]['weather-roost'][1]}/{len(stations[station]['weather-roost'][0])} & "\
        f"{stations[station]['AP-roost'][1]}/{len(stations[station]['AP-roost'][0])} & "\
        f"{stations[station]['unknown-noise-roost'][1]}/{len(stations[station]['unknown-noise-roost'][0])} & "\
        f"{stations[station]['duplicate'][1]}/{len(stations[station]['duplicate'][0])} & "\
        f"{stations[station]['bad-track'][1]}/{len(stations[station]['bad-track'][0])} & "\
        f"{stations[station]['non-roost'][1]}/{len(stations[station]['non-roost'][0])}"
    )
```

```python
# count pos --> neg and neg --> pos that happened during screening

import copy, json, os

DIR = "/scratch2/wenlongzhao/roostui/data/all_stations_v2_screened"
YEARS = range(2000, 2021)

columns = [
    "swallow-roost", "weather-roost", "AP-roost", "unknown-noise-roost",
    "roost -> roost", "non-roost -> roost",
    "duplicate", "bad-track", "non-roost",
    "roost -> non-roost", "non-roost -> non-roost",
]
high_conf_pred_to = {c: [0, 0] for c in columns} # (number of detections, number of tracks)
high_conf_pred_to["non-roost -> roost"] = ["-", "-"]
high_conf_pred_to["non-roost -> non-roost"] = ["-", "-"]
low_conf_pred_to = {c: [0, 0] for c in columns}
low_conf_pred_to["roost -> roost"] = ["-", "-"]
low_conf_pred_to["roost -> non-roost"] = ["-", "-"]

stations = {} # key: station, value: [file_names]
for file in sorted(os.listdir(DIR)):
    if not file.startswith("roost_labels"):
        continue
        
    station = file.split("_")[2]
    year = file.split("_")[3][:4]
    assert int(year) in YEARS
    if station not in stations:
        stations[station] = [file]
    else:
        stations[station].append(file)

statistics = {"ALL": None}
for station in stations: statistics[station] = None

for station in statistics:
    statistics[station] = {
        "pre-dualpol": {
            "high_conf_pred_to": copy.deepcopy(high_conf_pred_to),
            "low_conf_pred_to": copy.deepcopy(low_conf_pred_to),
        },
        "with-dualpol": {
            "high_conf_pred_to": copy.deepcopy(high_conf_pred_to),
            "low_conf_pred_to": copy.deepcopy(low_conf_pred_to),
        },
        "pre-dualpol & with-dualpol": {
            "high_conf_pred_to": copy.deepcopy(high_conf_pred_to),
            "low_conf_pred_to": copy.deepcopy(low_conf_pred_to),
        },
    } # statistics[station]["pre-dualpol"]["high_conf_pred_to"]["swallow-roost"]

# Finished init above, start the counting below
for station in stations:
    dualpol_info = json.load(open(f"../../wsrdata/tools/prepare_dataset_v0.2.0_help/all_days_all_scans_{station}.json"))
    
    for file in sorted(stations[station]):
        annotations = [annotation.strip().split(",") for annotation in open(os.path.join(DIR, file), "r").readlines()[1:]]
        tracks = {}
        for annotation in annotations:
            if int(annotation[10][4:6]) < 6 or int(annotation[10][4:6]) > 10: # check local month
                continue
            
            if annotation[0] not in tracks: # track_id
                dualpol = dualpol_info[f"{station}_{annotation[12][:4]}"]["all_scans_with_check"][annotation[1]]["dualpol"]
                tracks[annotation[0]] = [0, 0, 0, False, None, dualpol]
                    # sum score, num dets have score, num dets, has det w/ 0.5 score, screened as, dualpol

            # system predicted
            annotation[3] = float(annotation[3]) # det_score
            if annotation[3] > 0 and annotation[3] < 1:
                tracks[annotation[0]][0] = tracks[annotation[0]][0] + annotation[3] # sum score
                tracks[annotation[0]][1] = tracks[annotation[0]][1] + 1 # num dets have score
            tracks[annotation[0]][2] = tracks[annotation[0]][2] + 1 # num dets
            if annotation[3] >= 0.5: # has det w/ 0.5 score
                tracks[annotation[0]][3] = True
            
            # human screened
            if annotation[23].lower() in ["weather", "ap"]:
                tracks[annotation[0]][4] = "non-roost"
            else:
                tracks[annotation[0]][4] = annotation[20]

        for track_id, track in tracks.items():
            dualpol = "with-dualpol" if track[5] else "pre-dualpol"
            screened_category = track[4]
            if track[0] / track[1] >= 0.15 and track[2] >= 2 and track[3]: # high confidence prediction
                statistics[station][dualpol]["high_conf_pred_to"][screened_category][0] += track[2]
                statistics[station][dualpol]["high_conf_pred_to"][screened_category][1] += 1
                statistics["ALL"][dualpol]["high_conf_pred_to"][screened_category][0] += track[2]
                statistics["ALL"][dualpol]["high_conf_pred_to"][screened_category][1] += 1
            else: # low confidence predictiond
                statistics[station][dualpol]["low_conf_pred_to"][screened_category][0] += track[2]
                statistics[station][dualpol]["low_conf_pred_to"][screened_category][1] += 1
                statistics["ALL"][dualpol]["low_conf_pred_to"][screened_category][0] += track[2]
                statistics["ALL"][dualpol]["low_conf_pred_to"][screened_category][1] += 1

for station in statistics:
    for c in [
        "swallow-roost", "weather-roost", "AP-roost", "unknown-noise-roost",
        "duplicate", "bad-track", "non-roost",
    ]:
        for conf in ["high_conf_pred_to", "low_conf_pred_to"]:
            for i in [0, 1]:
                statistics[station]["pre-dualpol & with-dualpol"][conf][c][i] = \
                    statistics[station]["pre-dualpol"][conf][c][i] + \
                    statistics[station]["with-dualpol"][conf][c][i]
                    
    for dualpol in statistics[station]:
        for i in [0, 1]:
            statistics[station][dualpol]["high_conf_pred_to"]["roost -> roost"][i] = sum([
                statistics[station][dualpol]["high_conf_pred_to"][c][i]
                for c in ["swallow-roost", "weather-roost", "AP-roost", "unknown-noise-roost"]
            ])
            statistics[station][dualpol]["high_conf_pred_to"]["roost -> non-roost"][i] = sum([
                statistics[station][dualpol]["high_conf_pred_to"][c][i]
                for c in ["duplicate", "bad-track", "non-roost"]
            ])
            statistics[station][dualpol]["low_conf_pred_to"]["non-roost -> roost"][i] = sum([
                statistics[station][dualpol]["low_conf_pred_to"][c][i]
                for c in ["swallow-roost", "weather-roost", "AP-roost", "unknown-noise-roost"]
            ])
            statistics[station][dualpol]["low_conf_pred_to"]["non-roost -> non-roost"][i] = sum([
                statistics[station][dualpol]["low_conf_pred_to"][c][i]
                for c in ["duplicate", "bad-track", "non-roost"]
            ])
                
for station in statistics:
    for dualpol in statistics[station]:
        print(station, dualpol)
        print(",".join(["pre\\post screen"] + columns))
        print(",".join(["high confidence"] + [
            "/".join([
                str(statistics[station][dualpol]["high_conf_pred_to"][c][0]),
                str(statistics[station][dualpol]["high_conf_pred_to"][c][1]) 
            ]) for c in columns
        ]))
        print(",".join(["low confidence"] + [
            "/".join([
                str(statistics[station][dualpol]["low_conf_pred_to"][c][0]),
                str(statistics[station][dualpol]["low_conf_pred_to"][c][1]) 
            ]) for c in columns
        ]))
        if dualpol in ["pre-dualpol", "with-dualpol"]:
            print("----------")
        else:
            print("~~~~~~~~~~~~~~~~~~~~")
    
```
